```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
```

```python
# 좌표 데이터
X = np.array([[2, 3], [1, 3], [2, 2.5], [8, 3], [10, 3], [4, 4]])
#X = np.array([[2, 3], [9, 3], [6, 3]])
# x, y 컬럼 주고 위 데이터 프레임화
data = pd.DataFrame(X, columns=['x', 'y'])
# 데이터 정보 확인
data.info()
# plot으로 데이터 그려보기
data.plot(kind="scatter", x="x",y="y",figsize=(5,5),color="red")
```

```python
# kmeans 모델 생성 및 학습, 2개로 분류
kmeans = KMeans(n_clusters=2, random_state=0).fit(X)
kmeans.labels_
# [2,3]은 1번그룹 [1,3]은 1번그룹... 이런 식
```

```python
# 아래 문장에 오류 있으면 pip install threadpoolctl==3.1.0 후 restart the kernel
# 그럼 [0,0], [12,3]은 몇 번 그룹이냐
print(kmeans.predict( [[12, 3]] ))
print(kmeans.predict( [[0, 0] ]))
#kmeans.predict( [[0, 0], [12, 3]] )
```

```python
centers=kmeans.cluster_centers_
print(centers)
# [12, 3], [0,0]의 중간좌표 출력(아래 그림의 새모)
```

```python
# cluster center 그려보기
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1], marker='o')
plt.scatter(centers[:,0], centers[:,1], marker='^')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
