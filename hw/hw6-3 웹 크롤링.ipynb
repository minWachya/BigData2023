### 웹 크롤링 (Q6.3)
- 배운 내용을 활용하여 기상청의 오늘의 날씨를 크롤링해보자
- https://www.weather.go.kr/weather/observation/currentweather.jsp
- 지역별로 오늘의 온도와 습도를 크롤링하여 DataFrame을 만든다.
- [결과] ,sido-gu,온도,습도 0,강릉,3.5,29 1,강진군,3.0,39 2,강화,1.3,35 3,거제,3.7,26 4,거창,2.8,23 5,경주시,4.8,1...... 5507,홍성,1.4,35 5508,홍천,0.8,31 5509,흑산도,1.9,6

```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import datetime

from selenium import webdriver
import time

# 정적 웹 크롤링
# 크롤링
def get_weather(result):
    for page in range(1,59):
        weather_url = 'https://www.weather.go.kr/w/obs-climate/land/city-obs.do'
        
        html = urllib.request.urlopen(weather_url)
        soupHollys = BeautifulSoup(html, 'html.parser')
        
        # 지역명: tbody > tr > 0번째 td > a 
        # 온도: tbody > tr > 5번째 td 
        # 습도: tbody > tr > 10번째 td 
        tag_tbody = soupHollys.find('tbody')
        
        # 아래로 내려가서.. tbody 안의 모든 tr에 대해 실행
        for weather_info in tag_tbody.find_all('tr'):
            area = weather_info.find('a').string
            temperature = weather_info.find_all('td')[5].string
            humidity = weather_info.find_all('td')[10].string
            print(area)
            
            result.append([area]+[temperature]+[humidity])
    return


def main():
    result = []
    print('Weather crawling >>>>>>>>>>>>>>>>>>>>>>>>>>')
    get_weather(result)   
    w_tbl = pd.DataFrame(result, columns=('sido-gu', '온도', '습도')) # result를 dataframe으로 변경
    w_tbl.to_csv('./hw6-3-weather.csv', encoding='cp949', mode='w', index=True)
    del result[:]

if __name__ == '__main__':
     main()
```
