#### 2. 상관관계와 인과관계
- 인과관계: 선행하는 한 변인이 후행하는 다른 변인의 원인이 되고 있다고 믿어지는 관계

- 양의 상관관계: a 증가, b 증가
- 음의 상관관계: a 증가, b 감소
- 무 상관관계: a, b 증감 없음(두 변수간 증감 없음)

- 상관관계가 있다고 해도 인과관계 없는 경우가 있다....

- 상관관계: 수학적으로 증명이 가능
- 인과관계: 어디까지나 충분한 재현성의 확인, 변인의 배제, 통제집단과 실험집단의 설정과 같은 환경에서 얻어진 실험 데이터를 통해서 누적적으로 뒷받침될 뿐

- <어떤 상관관계가 확인되면, 잠재적 인과관계에 대해서는 다음과 같은 다섯 가지 경우가 있다.>
1. 처음부터 인과관계 같은 건 없었다. 단순한 우연의 일치였던 것.
2. 연구에 반영되지 않았던 제3의 변인 z 가 x 와 y 두 변인에게 영향을 끼칠 수 있다. 
3. 인과관계가 존재하기는 하는데, 서로가 서로에게 원인인 동시에 결과가 된다.
4. 인과관계가 존재하기는 하는데, x 가 원인이고 y 가 결과가 된다. 
5. 인과관계가 존재하기는 하는데, y 가 원인이고 x 가 결과가 된다.

- '상관관계는 인과관계와 같지 않다. 상관관계는 인과관계의 단순한 필요 조건의 하나이다'

#### 4. 회귀분석
(보통 이거를 많이 하는 듯)
- 종속변수가 1개, 독립 변수가 1개 -> 단 변량 단순 선형 회귀 모델 (Univariate simple linear regression model)
- 종속변수가 1개, 독립변수가 2개 이상 -> 단 변량 다중 선형 회귀 모델 (Univariate Multiple linear regression model)

(여기서부턴 잘 다루지 않음)
- 종속변수가 2개 이상, 독립변수가 1개 -> 다변량 단순 선형 회귀 모델 (Multivariate simple linear regression model)
- 종속변수가 2개 이상, 독립 변수가 2개 이상 -> 다변량 다중 선형 회귀 분석 (Multivariate multiple linear regression model)

## 1) 데이터 수집

```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston
boston =  load_boston()

# 함수 어떤 거 있는지 살펴보기
import sklearn
dir(sklearn.datasets)
```

## 2) 데이터 준비 및 탐색

```python
# boston 데이테셋의 description 
print(boston.DESCR)
```

```python
boston_df = pd.DataFrame(boston.data, columns = boston.feature_names)
boston_df.head()

# – CRIM: 지역별 범죄 발생률
# – ZN: 25,000평방피트를 초과하는 거주 지역 비율
# – INDUS: 비상업 지역의 넓이 비율
# – CHAS: 찰스강의 더미변수(1은 강의 경계, 0은 경계 아님)
# – NOX: 일산화질소 농도
# – RM: 거주할 수 있는 방 개수
# – AGE: 1940년 이전에 건축된 주택 비율
# – DIS: 5개 주요 고용센터까지 가중 거리
# – RAD: 고속도로 접근 용이도
# – TAX: 10,000달러당 재산세 비율
# – PTRATIO: 지역의 교사와 학생 수 비율
# – B:지역의흑인거주비율
# – LSTAT: 하위 계층의 비율
# – PRICE(MEDV): 본인 소유 주택 가격의 중앙값
```

```python
boston_df['PRICE'] = boston.target
boston_df.head()
```

```python
print('보스톤 주택 가격 데이터셋 크기 : ', boston_df.shape)
```

```python
# 각 속성별 타입
boston_df.info()

# 506개의 데이터가 있군. 타입은 다 플로트니 데이터 분석하기 쉽겠군
```

## 3) 분석 모델 구축

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# X, Y 분할하기
# Y = 결과값 == 집값
Y = boston_df['PRICE']
# X = 집값 제외 집값을 분석할 값들 / 수직으로 잘라내고 / boston_df를 수정하지 않고 결과만 X에 담기
X = boston_df.drop(['PRICE'], axis=1, inplace=False)
```

```python
# 훈련용 데이터와 평가용 데이터 분할하기
# 30%는 테스트용
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=156)
```

```python
# 선형회귀분석 : 모델 생성
lr = LinearRegression()
```

```python
# 선형회귀분석 : 모델 훈련
# 70% 가지고 모델 훈련
lr.fit(X_train, Y_train)
```

```python
# 선형회귀분석 : 평가 데이터에 대한 예측 수행 -> 예측 결과 Y_predict 구하기
Y_predict = lr.predict(X_test)
```

## 4) 결과 분석 및 시각화

```python
mse = mean_squared_error(Y_test, Y_predict)
rmse = np.sqrt(mse)

print('MSE : {0:.3f}, RMSE : {1:.3f}'.format(mse, rmse))
# 실제값과 예측값 비교
print('R^2(Variance score) : {0:.3f}'.format(r2_score(Y_test, Y_predict))) # 0.757면 나쁘진 않은 값. 설명력이 75%
```

```python
print('Y 절편 값: ', lr.intercept_)
print('회귀 계수 값: ', np.round(lr.coef_, 1))
```

```python
coef = pd.Series(data = np.round(lr.coef_, 2), index=X.columns)
coef.sort_values(ascending = False)

# 어느게 더 영향이 있는지는 이것만 봐선 모름
```
