```python
# 2019년1월 지하철 유임/무임 승하차, 무임승차 비율이 높은역 찾아 나가기
import csv
f = open(subwayfee_path, 'r', encoding='cp949')
data = csv.reader(f)
next(data)
mx = 0
rate = 0
for row in data :
    for i in range(4,8) :
        row[i] = int(row[i])
    # 분모가 0 아닐 때만 계산
    if row[6] != 0 :
        rate = row[4] / row[6]
        if rate > mx :
            mx = rate
            print(row, round(rate,2))
f.close()
```

```python
# 2019년1월 지하철 유임/무임 승하차, 유임승차 비율, 100000만 이상 승차역
import csv
f = open(subwayfee_path, 'r', encoding='cp949')
data = csv.reader(f)
next(data)
mx = 0
rate = 0
for row in data :
    for i in range(4,8) :
        row[i] = int(row[i]) # 정수처리
    # 분자 0 아니고 유무임 승차한 사람의 합이 많을 때만 계산
    if row[6] !=0 and (row[4] + row[6]) >100000 :
        rate = row[4] / (row[4] + row[6])
        if rate > mx :
            mx = rate
            maxrow=row; maxrate=rate
print(maxrow, round(maxrate,2))
f.close()
```

```python
# 2019년1월 지하철 유임/무임 승하차, 가장 많은역 이름
import csv
f = open(subwayfee_path, 'r', encoding='cp949')
data = csv.reader(f)
next(data)
mx = [0] * 4          # max값도 4개인 배열
mx_station = [''] * 4 # 문자열 길이 4인 배열

for row in data :
    # 4번 반복
    for i in range(4,8) :
        row[i] = int(row[i])
        # 4가지 값에 대한 최댓값 찾기
        # 역 이름 기억: 역 이름 + 호선
        if row[i] > mx[i-4] :
            mx[i-4] = row[i]
            mx_station[i-4] = row[3] + ' ' + row[1]
            
for i in range(4) :
    print(mx_station[i], mx[i])
f.close()
```

```python
# 2019년1월 지하철 유임/무임 승하차, 가장 많은역 이름
# 위에서 레이블만 추가
import csv
f = open(subwayfee_path, 'r', encoding='cp949')
data = csv.reader(f)
next(data)
mx = [0] * 4
mx_station = [''] * 4
label = ['유임승차','유임하차','무임승차','무임하차']
for row in data :
    for i in range(4,8) :
        row[i] = int(row[i])
        if row[i] > mx[i-4] :
            mx[i-4] = row[i]
            mx_station[i-4] = row[3] + ' ' + row[1]
for i in range(4) :
    print(label[i] + ' : ' + mx_station[i], mx[i])
```

```python
import csv
import matplotlib.pyplot as plt
f = open(subwayfee_path, 'r', encoding='cp949')
data = csv.reader(f)
next(data)
label = ['유임승차','유임하차','무임승차','무임하차']
c = ['#14CCC0', '#389993', '#FF1C6A', '#CC14AF']
plt.rcParams['font.family'] = 'AppleGothic'
list=['서울역', '시청', '건대입구']
for row in data :
    if row[3] not in list:
        continue
    for i in range(4,8) :
        row[i] = int(row[i])
    #plt.figure(dpi = 200)
    plt.title(row[3] + ' ' + row[1])
    plt.pie(row[4:8], labels = label, colors = c, autopct = '%1.f%%')
    plt.axis('equal')
    #plt.savefig(row[3] + ' ' + row[1] + '.png')  # 그림파일 생성
    plt.show()
f.close()
```
